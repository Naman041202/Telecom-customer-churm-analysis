```python
df['TotalCharges'] = df['TotalCharges'].replace(" ",'0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
df.info()
```

```python
 df.duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x ='Churn',data =df)
plt.title('Count of customers by Churn ')

ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb = df.groupby('Churn').agg({'Churn': 'count'})
plt.pie(gb['Churn'],labels= gb.index,autopct= "%1.2f%%")
plt.title('percentage of customers by Churn')
plt.show()
```

```python
plt.figure(figsize=(4,4))
gp = sns.countplot(x ='gender',data = df,hue = 'Churn')
gp.bar_label(gp.containers[0])
plt.title('Churn by gender')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sgp= sns.countplot(x = "SeniorCitizen", data = df,hue = 'Churn')
sgp.bar_label(sgp.containers[0])
plt.title('Churn by SeniorCitizen')
plt.show()
```

```python
 # Step 1: Create cross-tab (counts)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert to percentage
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar
ax = ct_pct.plot(kind='bar', stacked=True, figsize=(5,5))

# Step 4: Add % labels
for i in range(len(ct_pct)):
    cumulative = 0
    for j in range(len(ct_pct.columns)):
        value = ct_pct.iloc[i, j]
        ax.text(i, cumulative + value/2, f'{value:.1f}%', 
                ha='center', va='center', fontsize=9)
        cumulative += value

# Step 5: Titles & labels
plt.title('Churn by SeniorCitizen (%)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')

plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(x='tenure',data =df,bins =50,hue = "Churn")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sgp= sns.countplot(x = "Contract", data = df,hue = 'Churn')
sgp.bar_label(sgp.containers[0])
plt.title('Churn by Contracts')
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

for col in cols:
    plt.figure(figsize=(4,4))
    
    ax = sns.countplot(x=col, data=df, hue='Churn')
    
    # Add labels on all bars
    for container in ax.containers:
        ax.bar_label(container, padding=2)
    
    plt.title(f'Churn by {col}')
    plt.show()
```
